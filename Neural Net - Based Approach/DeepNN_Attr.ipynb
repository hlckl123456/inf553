````
from keras.models import Model
from keras.layers import Input, Dense

class NN():
 
  def __init__(self,train_business,alpha,output_dim):
        
        # the shape of one training example is

        self.alpha = alpha
        business_inp, business_model = self.createModel(train_business)
   
        if output_dim>1:
                  output_layer = Dense(output_dim,activation='softmax')(business_model)
                  self.model = Model(inputs=business_inp,outputs=output_layer)
                  self.model.compile(optimizer=keras.optimizers.Adam(alpha), loss='categorical_crossentropy', metrics=['accuracy'])
        else:
                  output_layer = Dense(output_dim,activation='sigmoid')(business_model)
                  self.model = Model(inputs=business_inp,outputs=output_layer)
                  self.model.compile(optimizer=keras.optimizers.Adam(alpha), loss='mse', metrics=['accuracy'])
        
        
        
  def createModel(self,training_data):
        input_shape = training_data.shape[1]
        inp_layer = Input(shape=(input_shape,))
        dense_layer = Dense(units=64, activation='relu')(inp_layer)
        dense_layer = Dense(units=32, activation='relu')(dense_layer)

        #model.add(Dense(units=5, activation='softmax'))
        return inp_layer, dense_layer
   
  def train(self,train_business,output_rating,num_iter):
    
        
        self.model.fit(train_business, output_rating, epochs=num_iter)
  
 
  
      
````
